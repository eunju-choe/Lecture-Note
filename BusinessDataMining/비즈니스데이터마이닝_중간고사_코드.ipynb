```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 문제 1-1

```python
# 데이터를 불러와 df라는 변수에 저장한다.
df = pd.read_csv('lendingclub.csv')

# describe함수를 사용하여 수치형 변수의 통계값을 분석한다.
df.describe()
```

# 문제 1-2

```python
# grade 변수의 종류를 파악하기 위해 describe함수를 사용한다.
df['grade'].describe()
```

```python
# 범주형 변수의 그래프를 그리기 위하여 value_counts함수를 사용한다.
# sort_index를 사용하여 값을 인덱스를 기준으로 정렬한다.
g = df['grade'].value_counts().sort_index()
print(g)
```

```python
# 비중을 비교하기 위해서 pie 그래프를 그린다.
plt.pie(g, labels = [0, 1, 2, 3, 4, 5, 6], autopct = '%1.2f%%' );
```

# 문제 1-3

```python
# corr함수를 사용하여 전체 데이터의 상관관계 표를 그리고, cor 변수에 저장한다.
cor = df.corr()
cor
```

```python
# grade의 corr만 추출하여 상관관계가 가장 높은 변수를 찾는다.
cor['grade']
```

# 문제 1-4

```python
# 결측치를 찾기 위해 isna함수를 사용하고, 결측치의 개수를 구하기 위해 sum 함수를 사용한다.
df['mths_since_recent_inq'].isna().sum()
```

```python
# 결측치 처리 전의 통계량
df['mths_since_recent_inq'].describe()
```

```python
# 해당 변수의 평균을 구하여 mean 변수에 저장한다.
mean = df['mths_since_recent_inq'].mean()

# 결측치에 임의로 mean을 채워넣는다.
df["mths_since_recent_inq"]=df["mths_since_recent_inq"].fillna(mean)
```

```python
# 결측치 처리가 된 것을 확인하였다.
df['mths_since_recent_inq'].isna().sum()
```

```python
# 결측치 처리 후의 통계량
df['mths_since_recent_inq'].describe()
```
