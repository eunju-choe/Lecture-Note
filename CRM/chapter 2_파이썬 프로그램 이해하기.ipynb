## 주요 list 멤버 함수(Method)

```python
a = [1, 2, 3, 4, 5]
```

```python
# append : 리스트의 끝에 항목을 추가함
a.append(6)
print(a)
```

```python
# insert : 리스트의 원하는 위치에 값을 추가
a.insert(2, 2.5)
print(a)
```

```python
# remove : 값으로 항목 삭제
a.remove(2.5)
print(a)
```

```python
# index : 찾고자 하는 항목의 인덱스 반환
a.index(5)
```

```python
# in : 리스트 내에 해당 항목이 존재하는지 확인
print(7 in a)
print(1 in a)
```

```python
# sort : 리스트 자체를 내부적으로 정렬
a = [4, 5, 90, 1, 6, 87, 6]
a.sort()
print(a)
```

# 함수(function)

```python
# 내장함수
a = [1, 2, 3, 4, 5]

print(max(a))
print(min(a))
print(sum(a))
print(len(a))
```

```python
# 사용자 정의 함수
def arith_oper(n, m) :
    a = n + m
    b = n - m 
    c = n * m
    d = n / m
    e = n % m
    
    return a, b, c, d, e


arith_oper(3, 4)
```

```python
# 람다(lambda) 함수
arith_oper_1 = lambda n, m : (n+m, n-m, n*m, n/m, n%m)

arith_oper_1(3, 4)
```

# if문

```python
# if-else
x = 10

if x == 10:
    print('철수')
    print('영희')
else:
    print('대한')
    print('민국')
```

```python
# if-elif-else
score = float(input('성적을 입력하시오.'))

if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
elif score >= 60:
    print('D')
else:
    print('F')
```

# 반복문

```python
hap = 0

for i in range(1, 11):
    hap += i
    
print('합계 :', hap)
```

```python
hap = 0
for i in range(100, 1001):
    hap += 1
    if hap >= 500 :
        break

print('조건 만족')
print(i, hap)
```

```python
# 사용자 정의 함수를 이용하여 구구단 출력하기
def gugudan(num):
    if num >= 10 : 
        print('1부터 9 사이의 값을 입력하세요.')
    
    else:
        for i in range(1, 10):
            print(f'{num} * {i} = {num*i}')
        
num = int(input('숫자를 입력하세요 : '))
gugudan(num)
```
