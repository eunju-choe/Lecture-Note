```python
df_new = df.dropna()
df_new.head()
```

```python
df_new.describe()
```

```python
df_new = df_new[df_new['Quantity']>0]
df_new.info()
```

```python
df.describe()
```

```python
df_new['Country'].value_counts()
```

```python
# 형식 변경
df_new['CustomerID'] = df_new['CustomerID'].astype('int64')
df_new['InvoiceDate'] = df_new['InvoiceDate'].astype('datetime64')
df_new.info()
```

```python
df_new['TotalPrice'] = df_new['Quantity'] * df_new['UnitPrice']
df_new.describe()
```

```python
df_new.head()
```

```python
# 컬럼 순서 변경
df_new = df_new[['InvoiceNo', 'InvoiceDate', 'StockCode', 'Description', 'Quantity', 'UnitPrice', 'TotalPrice', 'CustomerID', 'Country']]
df_new.head()
```

```python
# 컬럼 이름 변경
df_new.rename(columns = {'TotalPrice' : 'Total', 'UnitPrice' : 'Unit'}, inplace=True)
df_new.head()
```

```python
# 열 이름 변경
# 'StockCode' -> 'Code'

df_new = df_new.rename(columns={'StockCode' : 'Code'})
df_new.head()
```

```python
df_new.to_csv('data/OnlineRetail_new.csv', index=False)
```
