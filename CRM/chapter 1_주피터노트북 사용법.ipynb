# Python 기본 연산자

## 산술 연산자

```python
num1 = 100
num2 = 50

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 % num2)
print(num1 ** num2)
```

## 비교 연산자

```python
num1 = 100
num2 = 50

print(num1 == num2)
print(num1 !=num2)
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
```
